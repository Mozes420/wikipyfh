```python
df = pd.read_csv('enwiki-p1p857.csv', quotechar='|', sep = '#', engine = 'python', on_bad_lines='warn')
df['timestamp'] = pd.to_datetime(df['timestamp'],format='%Y-%m-%dT%H:%M:%SZ')
```

```python
df.head()
```

```python
rid.clear()
txt.clear()
```

```python
import mwxml
import glob

paths = glob.glob('enwiki-20220201-pages-meta-history1.xml-p1p857.bz2')

def process_dump(dump, path):
  for page in dump:
    for revision in page:
        yield revision.id, revision.text

count = 0
rid = []
txt = []

for rev_id, rev_text in mwxml.map(process_dump, paths):
  rid.append(rev_id)
  txt.append(rev_text)
  count += 1
  if count > 15:
    break
```

```python
for e in txt:
    e.replace('\n', ' ')
```

```python
ridtxt = list(zip(rid, txt))
```

```python
dftxt = pd.DataFrame(ridtxt, columns = ['revision_id', 'revision_text'])
dftxt['revision_text'] = dftxt['revision_text'].str.replace('\n',' ', regex = True)
dftxt
```

```python
df_outer = pd.merge(df, dftxt, on='revision_id', how='outer')
df_outer.head(20)
```
